```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline

from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import GradientBoostingRegressor, BaggingRegressor

from sklearn.metrics import mean_squared_error, r2_score
```

```python
#Loading dataset
#Training dataset_sales
sales = pd.read_csv("sales.csv")
```

```python
#feature Engg
#converting the date
sales["Date"] = pd.to_datetime(sales["Date"])
```

```python
sales["Year"] = sales["Date"].dt.year
sales["Month"] = sales["Date"].dt.month
sales["Day"] = sales["Date"].dt.day
sales["Week"] = sales["Date"].dt.isocalendar().week.astype(int)
```

```python
sales.drop("Date", axis=1, inplace=True)
```

```python
sales.head()
```

```python
#seperating Features and target
X = sales.drop("Sales", axis=1)

y = sales["Sales"]
```

```python
#Identify the categorical Feature
categorical_features = ["State_holiday"]
```

```python
#One-Hot Encoding
preprocessor = ColumnTransformer(
    transformers=[
        (
            "holiday",
            OneHotEncoder(handle_unknown="ignore"),
            categorical_features
        )
    ],
    remainder="passthrough"
)
```

```python
#Spliting dataset(Test,train)
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)
```

```python
#Creating model Pipelines
#Linear regression(Training)
linear_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", LinearRegression())
])
```

```python
linear_pipeline.fit(X_train, y_train)
```

```python
#Decision Tree(Training)
tree_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", DecisionTreeRegressor(random_state=42))
])
```

```python
tree_pipeline.fit(X_train, y_train)
```

```python
#Linear Regression(Evaluating)
linear_predictions = linear_pipeline.predict(X_test)

mse_linear = mean_squared_error(y_test, linear_predictions)
rmse_linear = np.sqrt(mse_linear)
r2_linear = r2_score(y_test, linear_predictions)

print("Linear Regression")
print("MSE :", mse_linear)
print("RMSE:", rmse_linear)
print("R² :", r2_linear)
```

```python
#decision Tree(Evaluating)
tree_predictions = tree_pipeline.predict(X_test)

mse_tree = mean_squared_error(y_test, tree_predictions)
rmse_tree = np.sqrt(mse_tree)
r2_tree = r2_score(y_test, tree_predictions)

print("Decision Tree")
print("MSE :", mse_tree)
print("RMSE:", rmse_tree)
print("R² :", r2_tree)
```

 # RandomForestRegressor

```python

from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline

# Random Forest (Training)
rf_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("model", RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1))
])

rf_pipeline.fit(X_train, y_train)

# Random Forest (Evaluating)
rf_predictions = rf_pipeline.predict(X_test)

mse_rf = mean_squared_error(y_test, rf_predictions)
rmse_rf = np.sqrt(mse_rf)
r2_rf = r2_score(y_test, rf_predictions)

print("Random Forest")
print("MSE :", mse_rf)
print("RMSE:", rmse_rf)
print("R² :", r2_rf)
```

```python
results = pd.DataFrame({
    "Model": ["Linear Regression", "Decision Tree", "Random Forest"],
    "RMSE": [rmse_linear, rmse_tree, rmse_rf],
    "R2": [r2_linear, r2_tree, r2_rf]
})
print(results)
```
